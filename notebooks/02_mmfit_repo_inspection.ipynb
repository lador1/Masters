```python
from pathlib import Path

REPO_DIR = Path(r"C:\Masters\external\mm-fit-code")

search_terms = [
    "joint",
    "joints",
    "skeleton",
    "pose_3d",
    "keypoint",
    "keypoints",
    "openpose",
    "COCO",
    "pose",
]

matches = []

for path in REPO_DIR.rglob("*"):
    if path.is_file() and path.suffix.lower() in [".py", ".ipynb", ".md", ".txt", ".yml"]:
        try:
            text = path.read_text(encoding="utf-8", errors="ignore")
        except Exception:
            continue

        for term in search_terms:
            if term.lower() in text.lower():
                matches.append((path.relative_to(REPO_DIR), term))
                break

print("Files with possible pose/joint references:")
for file, term in matches:
    print(f"{file}  <-- matched: {term}")
```

```python
from pathlib import Path

REPO_DIR = Path(r"C:\Masters\external\mm-fit-code")

terms = ["joint", "skeleton", "pose_3d", "keypoint", "openpose", "COCO"]

for path in REPO_DIR.rglob("*"):
    if path.is_file() and path.suffix.lower() in [".py", ".md", ".txt", ".yml"]:
        try:
            lines = path.read_text(encoding="utf-8", errors="ignore").splitlines()
        except Exception:
            continue

        hits = []
        for i, line in enumerate(lines):
            if any(term.lower() in line.lower() for term in terms):
                hits.append((i + 1, line))

        if hits:
            print("\n" + "=" * 80)
            print(path.relative_to(REPO_DIR))
            print("=" * 80)
            for line_no, line in hits[:50]:
                print(f"{line_no}: {line}")
```
